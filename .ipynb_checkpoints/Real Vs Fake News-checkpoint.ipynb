```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
true = pd.read_csv("Data/True.csv")
fake = pd.read_csv("Data/Fake.csv")
true["fake"] = true.shape[0] * [0]
fake["fake"] = fake.shape[0] * [1]
```

```python
data = pd.concat([true,fake]).reset_index().drop("index",axis= 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data["text"], data["fake"], random_state= 1)
```

```python
vectorizer=TfidfVectorizer(stop_words='english')
vectorizer_train=vectorizer.fit_transform(X_train) 
vectorizer_test=vectorizer.transform(X_test)
```

```python
vectorizer_train
```

```python
RFC = RandomForestClassifier()
RFC.fit(vectorizer_train, y_train)
pred = RFC.predict(vectorizer_test)
```

```python
score=accuracy_score(y_test,pred)
```
